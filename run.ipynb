```python
import numpy as np
import mujoco
import mediapy as media

xml = 'low_cost_robot/scene.xml'
model = mujoco.MjModel.from_xml_path(xml)
data = mujoco.MjData(model)

model.vis.global_.offheight = 480
model.vis.global_.offwidth = 640

renderer = mujoco.Renderer(model, height=480, width=640)

camera = mujoco.MjvCamera()
mujoco.mjv_defaultFreeCamera(model, camera)
camera.distance = 1.0
```

```python
toe_angles = np.load('aist_angles/toe_angles.npy')
shoulder_angles = np.load('aist_angles/shoulder_angles.npy')
lower_body_angles = np.load('aist_angles/lower_body_angles.npy')
upper_body_angles = np.load('aist_angles/upper_body_angles.npy')
```

```python
lower_body_angles = lower_body_angles - np.pi/2
lower_body_angles = lower_body_angles*0.5
upper_body_angles = np.pi/2 - upper_body_angles
```

```python
frames = []
mujoco.mj_resetDataKeyframe(model, data, 1)

for toe_angle,shoulder_angle, lower_body_angle, upper_body_angle in zip(toe_angles, shoulder_angles,lower_body_angles,upper_body_angles):
    data.qpos[0] = toe_angle
    data.qpos[1] = lower_body_angle
    data.qpos[2] = upper_body_angle
    data.qpos[4] = shoulder_angle
    data.qpos[5] = np.pi/2
    mujoco.mj_forward(model, data) # forward kinematics
    renderer.update_scene(data, camera)
    frames.append(renderer.render())
```

```python
media.show_video(frames, fps=30, loop=False)
```
